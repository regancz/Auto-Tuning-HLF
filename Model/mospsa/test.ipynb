```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# 从数据库中读取数据
config_df = pd.read_sql('SELECT id, Orderer_BatchSize_MaxMessageCount FROM config_parameter_spsa', con=engine)
performance_df = pd.read_sql("SELECT config_id, throughput, bench_config FROM performance_metric_spsa WHERE succ <> 0", con=engine)

# 将两个表格通过config_id进行合并
merged_df = pd.merge(config_df, performance_df, left_on='id', right_on='config_id')

# 按照Orderer_BatchSize_MaxMessageCount的值进行排序
merged_df_sorted = merged_df.sort_values('Orderer_BatchSize_MaxMessageCount')
import seaborn as sns
sns.set(style='ticks')
plt.rcParams['font.sans-serif'] = 'SimHei'
# 创建子图
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
i = 0
# 分别绘制每个bench_config的折线图
for bench_config in ['open', 'transfer']:
    if bench_config == 'query':
        continue
    subset_df = merged_df_sorted[merged_df_sorted['bench_config'] == bench_config]
    subset_df = subset_df.drop(['id', 'config_id', 'bench_config'], axis=1)
    subset_df = subset_df.groupby('Orderer_BatchSize_MaxMessageCount').mean().reset_index()

    # 在对应的子图中绘制折线图
    axes[i].plot(subset_df['Orderer_BatchSize_MaxMessageCount'], subset_df['throughput'])
    axes[i].set_xlabel('MaxMessageCount')
    axes[i].set_ylabel('吞吐量')
    axes[i].set_title(f'{bench_config}'.capitalize())

    # 多项式拟合
    degree = 5  # 选择拟合的多项式次数
    coeffs = np.polyfit(subset_df['Orderer_BatchSize_MaxMessageCount'], subset_df['throughput'], degree)
    poly = np.poly1d(coeffs)

    # 在对应的子图中绘制拟合曲线
    axes[i].plot(subset_df['Orderer_BatchSize_MaxMessageCount'], poly(subset_df['Orderer_BatchSize_MaxMessageCount']),
                 label=f'Fit (Degree {degree})', linestyle='dashed')
    i += 1

# 调整布局
plt.tight_layout()
plt.show()
```
